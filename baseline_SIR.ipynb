```python
def hsl_to_rgb(h, s, l):
    h /= 360
    s /= 100
    l /= 100
    return colorsys.hls_to_rgb(h, l, s)

def draw_sir(all_f_res,net):
    hsl_lightcoral = (5, 90, 78)  # navajowhite in HSL
    rgb_lightcoral = hsl_to_rgb(*hsl_lightcoral)
    
    hsl_skyblue = (192, 80, 83)  # navajowhite in HSL
    rgb_skyblue = hsl_to_rgb(*hsl_skyblue)
    
    
    all_f_res_list = all_f_res[2]
    all_f_res_list_greedy_degree = all_f_res[1]
    all_f_res_list_random = all_f_res[0]
    with plt.style.context(['default']):
        fig, ax = plt.subplots(1,1,figsize=(5,3.5))
            
        for i in range(all_f_res_list.shape[0]):
                
            
            ax.plot(all_f_res_list_greedy_degree[i][0][:50], all_f_res_list_greedy_degree[i][2][:50],'-',alpha=0.15, color=rgb_lightcoral)
            ax.plot(all_f_res_list_random[i][0][:50], all_f_res_list_random[i][2][:50],'-',alpha=0.15, color=rgb_skyblue)
                    
        
        
        f_res_list_greedy_degree = np.mean(all_f_res_list_greedy_degree,axis=0)
        f_res_list_random = np.mean(all_f_res_list_random,axis=0)
            
        
        ax.plot(f_res_list_random[0][:50], f_res_list_random[2][:50],'-',label = r'Random selection',color='deepskyblue',linewidth=2.5)
        ax.plot(f_res_list_greedy_degree[0][:50], f_res_list_greedy_degree[2][:50],'-',label = r'Degree-greedy',color='r',linewidth=2.5)
        
        ax.legend(fontsize=6,loc=1,frameon=False)
        ax.set_ylim(-0.05, 1)
        ax.set_yticks(np.arange(0, 1.0, 0.2))
        ax.set_title(net, fontsize=10)
            
    with plt.style.context(['default']):
        fig, ax = plt.subplots(1,1,figsize=(5,3.5))
            
        for i in range(all_f_res_list.shape[0]):
                
            
            ax.plot(all_f_res_list_greedy_degree[i][0][:80], all_f_res_list_greedy_degree[i][3][:80],'-',alpha=0.15, color=rgb_lightcoral)
            ax.plot(all_f_res_list_random[i][0][:80], all_f_res_list_random[i][3][:80],'-',alpha=0.15, color=rgb_skyblue)
                    
        
        # f_res_list = np.mean(all_f_res_list,axis=0)
        f_res_list_greedy_degree = np.mean(all_f_res_list_greedy_degree,axis=0)
        f_res_list_random = np.mean(all_f_res_list_random,axis=0)
            
        
        ax.plot(f_res_list_random[0][:80], f_res_list_random[3][:80],'-',label = r'Random selection',color='deepskyblue',linewidth=2.5)
        ax.plot(f_res_list_greedy_degree[0][:80], f_res_list_greedy_degree[3][:80],'-',label = r'Degree-greedy',color='r',linewidth=2.5)
        
        ax.legend(fontsize=6,loc=0,frameon=False)
        ax.set_ylim(-0.05, 1.03)
        ax.set_yticks(np.arange(0, 1.0, 0.2))
        ax.set_title(net, fontsize=10)          
```

```python
# +
nets = ['_er','_ba_3','_sw_6_01','_4_3_065','_sf_4_3_05']
for net in nets:
    all_f_res = np.load('./results/tem_sir/sir'+net+'.npy')
    draw_sir(all_f_res,net)
```
